```python

# store target variable in target variable and drop it from the spectf dataframe
Outcome=airqd_norm_df['RH']
airqd_norm_df.drop('RH',axis=1,inplace=True)
airqd_norm_df_new=airqd_norm_df

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(airqd_norm_df_new,Outcome,test_size=0.3)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Load libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create decision tree classifer object
regr = RandomForestRegressor(random_state=0, n_jobs=-1)

# Train model
model = regr.fit(airqd_norm_df_new, Outcome)
```

```python
# Calculate feature importances
importances = model.feature_importances_
print(importances)
```

```python
# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [airqd_norm_df_new.columns[i] for i in indices]

# Create plot
plt.figure()

# Create plot title
plt.title("Feature Importance")

# Add bars
plt.bar(range(airqd_norm_df_new.shape[1]), importances[indices])
#train_df_new_new=plt.bar(range(train_df_new.shape[1]), importances[indices])
# Add feature names as x-axis labels
plt.xticks(range(airqd_norm_df_new.shape[1]), names, rotation=90)

# Show plot
plt.show()
```

```python
rfReg=RandomForestRegressor(max_depth = 5, min_samples_split=5, n_estimators = 15, random_state = 1)

# Train the Classifier to take the training features and learn how they relate
# to the training y (the species)
rfReg.fit(airqd_norm_df_new, Outcome)
```

```python
rfReg.score(airqd_norm_df_new,Outcome)
```

```python
rfReg.score(x_train,y_train)
```

```python
# Now test the model

# First transform theinput for the test
test_pred = rfReg.predict(x_test)
a=test_pred
b=y_test
result = pd.DataFrame(a,b)
result

```

```python
#Now we can see this model score on test data and we can say that this works well 

from sklearn import metrics
import numpy as np
print ('MAE:', metrics.mean_absolute_error(y_test, test_pred))
print ('MSE:', metrics.mean_squared_error(y_test, test_pred))
print ('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, test_pred)))
```
